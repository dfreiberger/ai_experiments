## OpenAI Completion API
Experimenting with openai's [Completion API](https://platform.openai.com/docs/api-reference/completions). It uses this to try to execute code as quickly as possible from a text prompt

## Prerequisite
You'll need to initialize an environment variable `OPENAI_API_KEY` in the code block below. Key available at https://platform.openai.com/account/api-keys 

# !!WARNING: Do not check in your openai API Key on github!!



````python
import os

## PREREQUISITE: Set env variable with the openai API key available at https://platform.openai.com/account/api-keys" 
os.environ['OPENAI_API_KEY'] = 'OPENAI_API_KEY_HERE'
````

````python
import openai
import traceback

openai.api_key = os.getenv('OPENAI_API_KEY')

def do_this(prompt, show_result=False, show_debugging_info=False):
    result = None
    try:
        completion = openai.Completion.create(
            engine="text-davinci-003",
            prompt="\n# ".join(prompt.split("\n")),
            echo=True,
            temperature=0,
            max_tokens=1024,
            top_p=0.1,
            frequency_penalty=0,
            presence_penalty=0
        )
        result = completion.choices[0].text
        if show_result:
            print(result)
        exec(result)
    except Exception as e:
        print("Program error : ", str(e))
        if show_debugging_info : 
            traceback.print_exc()
    return result
````

````python
prompt = f"""
Write a function in python that returns position as a function of time, with inputs for acceleration and initial velocity
Use pandas to show a plot using this function over a time span of 10 seconds, with an initial velocity of 10 m/s and an acceleration of -2 g
"""

result = do_this(prompt)
````
