This notebook plays around with the openai "Completion" API. It uses this to try to execute code as quickly as possible from a text prompt

```python
import openai

def do_this(prompt, show_result=False):
    result = None
    try:
        completion = openai.Completion.create(
            engine="text-davinci-003",
            prompt="\n# ".join(prompt.split("\n")),
            echo=True,
            temperature=0,
            max_tokens=1024,
            top_p=0.1,
            frequency_penalty=0,
            presence_penalty=0
        )
        result = completion.choices[0].text
        if show_result:
            print(result)
        exec(result)
    except:
        print("Error attempting to run the following program")
        print(result)
    return result
```

```python
result = do_this("""
Write a function in python that returns position as a function of time, with inputs for acceleration and initial velocity
Use pandas to show a plot using this function over a time span of 10 seconds, with an initial velocity of 10 m/s and an acceleration of -2 g
""")
```
